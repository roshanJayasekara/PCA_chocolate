```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("candy-data.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.iloc[:,1:-3]
```

```python
subset= df.iloc[:,1:-3]
subset.head()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(subset)
```

```python
pca.explained_variance_ratio_
```

```python
candy_2d = pd.DataFrame(pca.transform(subset))
candy_2d.head()
```

```python
candy_2d.plot(kind="scatter",x=0,y=1)
```

```python
df.head()
```

```python
np.random.randn(85)
```

```python
df["X_pca"] = candy_2d["0"] 
df["Y_pca"] = candy_2d["1"] 
```

```python
df["x_jitter"] = candy_2d["0"] + np.random.randn(85)*.1
df["y_jitter"] = candy_2d["1"] + np.random.randn(85)*.1
```

```python
df.plot(kind="scatter",x="x_jitter",y="y_jitter")
```
